## Encoding Categorical Variables

```python
# Check unique values in the species column
print("Unique species:", df['species'].unique())
print("Data type of species column:", df['species'].dtype)
```

```python
# Label Encoding — converts text to numbers
# setosa → 0, versicolor → 1, virginica → 2

le = LabelEncoder()
df['species_encoded'] = le.fit_transform(df['species'])

# Let's see the result
print("Before and After Encoding:")
print(df[['species', 'species_encoded']].head(10))
```

```python
# X = input features (what the model learns FROM)
# y = target/label (what the model is trying to PREDICT)

X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species_encoded']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
print("\nFirst 5 rows of X:")
print(X.head())
```

```python
# StandardScaler makes all features have:
# mean = 0 and standard deviation = 1

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Let's see the difference
print("Before scaling (first 3 rows):")
print(X.head(3).values)

print("\nAfter scaling (first 3 rows):")
print(X_scaled[:3])
```

```python
# 80% data → model learns from this (training)
# 20% data → we test how well the model learned (testing)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, 
    test_size=0.2,      # 20% for testing
    random_state=42     # ensures same split every time you run
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("\nTraining samples:", len(X_train))
print("Testing samples:", len(X_test))
```
